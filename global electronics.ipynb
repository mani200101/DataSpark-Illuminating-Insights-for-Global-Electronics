```python
import pandas as pd
from sqlalchemy import create_engine

# Load your CSV file containing the data
df = pd.read_csv('m_c_s_p.csv')

df['Unit Cost USD'] = df['Unit Cost USD'].replace({'\$': '', ',': '', ' ': ''}, regex=True)
df['Unit Price USD'] = df['Unit Price USD'].replace({'\$': '', ',': '', ' ': ''}, regex=True)
df['Unit Cost USD'] = df['Unit Cost USD'].astype(float)
df['Unit Price USD'] = df['Unit Price USD'].astype(float)


# Convert 'Birthday' from DD/MM/YYYY to YYYY-MM-DD format
df['Birthday'] = pd.to_datetime(df['Birthday'], format='%d/%m/%Y').dt.strftime('%Y-%m-%d')

# Save the corrected data back to a CSV file (optional)
df.to_csv('data_cleaned.csv', index=False)

# Connect to MySQL database using SQLAlchemy
engine = create_engine("mysql+mysqlconnector://username:password@localhost/your_database")

# Insert corrected data into MySQL
df.to_sql('your_table', con=engine, if_exists='replace', index=False)

print("Data successfully imported with corrected date format!")

```
